```python
import numpy as np 
import random 
import matplotlib.pyplot as plt
np.random.seed(10)
```

```python
X_ = np.sort(np.random.uniform(low=0.0, high=1.0, size=500))
Y_ = np.sin(2*np.pi*X_)
noise_ = np.random.normal(scale=0.3, size=Y_.shape)
YY_ = Y_+noise_
```

```python
plt.plot(X_,YY_)
plt.ylabel("amplitude")
plt.xlabel("time")
plt.show()
```

```python
random.seed(10)
ind_ = random.sample(range(500), 500)
ind_train = ind_[:200]
X_sub_train = X_[ind_train]
YY_sub_train = YY_[ind_train]
```

```python
from sklearn.linear_model import LinearRegression
L_1 = LinearRegression()
L_1.fit(X_sub_train[:, np.newaxis], YY_sub_train)
predictions= L_1.predict(X_[:,np.newaxis])
plt.plot(X_, Y_, color = 'green')
plt.plot(X_,predictions)
plt.show()
```

```python
from sklearn.preprocessing import PolynomialFeatures
poly_ = PolynomialFeatures(degree=50)
X_sub_poly = poly_.fit_transform(X_sub_train[:, np.newaxis])
print(X_sub_poly.shape)

X_poly = poly_.fit_transform(X_[:,np.newaxis])
L_1.fit(X_sub_poly, YY_sub_train)
pred_ = L_1.predict(X_poly)
plt.plot(X_, Y_, color = 'green')
plt.plot(X_, pred_, color = 'red')
plt.ylim(-1.25,1.25)
plt.scatter(X_sub_train, YY_sub_train, s=50, facecolor='none', edgecolors='b')
plt.show()
```
